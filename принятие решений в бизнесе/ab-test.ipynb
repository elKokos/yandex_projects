##### convert_to_lowercase

```python
def convert_to_lowercase(data):
    
    """
    Функция приводит к нижнему регистру значения столбцов и строк из DataFrame,
    в наименованиях столбцов заменяет пробелы на андерскор,
    возвращает обновленный DataFrame
    Параметры:
    - data: исходные данные.
    """
    # приведение значения столбцов к нижнему регистру
    data.columns = data.columns.str.lower()
    # замена пробелов на андерскор
    data.columns = data.columns.str.replace(' ', '_')
    
    # работа со значениями в строчках только с типом данных object
    for column in data.select_dtypes(include='object'):
        data[column] = data[column].str.lower()
    
    # возвращаем обновленную таблицу
    return data
```

<a id='3_3_2'></a>
##### prioritize_hypotheses

```python
def prioritize_hypotheses(data, method='ice'):
    """
    Функция, приоритизирующая гипотезы методом ICE или RICE,
    создает необходимые столбцы (в зависимости от выбранного метода) 
    и сортирует по убыванию DataFrame по рангу набранных баллов.
    Возвращает обновленный DataFrame.
    Параметры:
    - data: исходные данные;
    - method: метод использующийся для приоретизации гипотез (по-умолчанию ICE).
    
    -------
    ВАЖНО!!! DataFrame должен содержать необходимые столбцы, а именно:
    - impact (влияние);
    - confidence (уверенность);
    - efforts (усилия);
    - reach (охват).
    """
    
    if method == 'ice':
        # считаем и округляем до 3 цифр после запятой
        data['ice_score'] = ((data.impact * data.confidence) / data.efforts).round(3)
        
        # ранжируем для наглядности + преобразуем ранк в целое число
        data['ice_rank'] = data['ice_score'].rank(ascending=False).astype(int)
        
        # возвращаем обновленный DataFrame с сортировкой по рангу
        return data.sort_values(by='ice_rank')
    
    elif method == 'rice':
        # считаем и округляем до 3 цифр после запятой
        data['rice_score'] = (data.reach * data.impact * data.confidence) / data.efforts
        
        # ранжируем для наглядности + преобразуем ранк в целое число
        data['rice_rank'] = data['rice_score'].rank(ascending=False).astype(int)
        
        # возвращаем обновленный DataFrame с сортировкой по рангу
        return data.sort_values(by='rice_rank')
    else:
        print('Ошибка! нет такого метода')
        return data
```

##### ordersAggregated

```python
# получаем агрегированные кумулятивные по дням данные о заказах
ordersAggregated = orders.groupby(['date', 'group']).agg({
    'transactionId': 'nunique',
    'visitorId': 'nunique',
    'revenue': 'sum'
}).groupby(level=1).cumsum().reset_index()

ordersAggregated.head(10)
```

##### visitorsAggregated

```python
# получаем агрегированные кумулятивные по дням данные о пользователях интернет-магазина
visitorsAggregated = (visitors.groupby(['date', 'group'])
                      .agg({'visitors': 'sum'})
                      .groupby(level=1).cumsum()
                      .reset_index())

visitorsAggregated.head(10)
```

##### cumulativeData

```python
# объединяем кумулятивные данные в одной таблице и присваиваем ее столбцам понятные названия
cumulativeData = ordersAggregated.merge(visitorsAggregated, on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
```

#### Подсчет статистической значимость различий в среднем количестве заказов на пользователя между группами по «сырым» данным. Выводы и предположения.

1. Сформулируем гипотезу:
    - Нулевая гипотеза (H0): Среднее количество заказов на пользователя из группы *a*  равно среднему количеству заказов на пользователя из группы *b*.
    - Альтернативная гипотеза (H1): Среднее количество заказов на пользователя из группы *a*  отличается от среднего количества заказов на пользователя из группы *b*.

2. Выполнить тест:
    - Проверим гипотезу о равенстве средних количеств заказов на пользователя из разных групп.
    - Воспользуемся Z-критерием для разности двух долей.
    - Получить p-value и сравнить с критическим уровенем статистической значимости.
    
3. Проанализировать результаты:
    - Если p-value меньше выбранного уровня значимости (alpha = 0.05), то отвергаем нулевую гипотезу и делаем вывод, что среднее количество заказов на пользователя из группы *a*  отличается от среднего количества заказов на пользователя из группы *b*.
    - Если p-value больше выбранного уровня значимости, то нулевая гипотеза не может быть отвергнута и нельзя сделать вывод об отличии среднего количества заказов на пользователя из группы *a* от среднего количества заказов на пользователя из группы *b*.

```python
#разбор кода:
# создаем таблицу с количеством заказов и количеством пользователей в разрезе групп
pivot_avg_orders = (orders.groupby('group')
                    # группируем данные orders по группа a и b
                    # считаем количество заказов
                    .agg({'transactionId':'count'})
                    # сбррасываем индексы
                    .reset_index()
                    # добавляем к получившейся таблице 
                    # сгруппированные данные из таблицы visitors по столбцу group
                    .merge(visitors.groupby('group')
                           # считаем количество пользователей
                           .agg({'visitors': 'sum'})
                           # сбрасываем индексы
                           .reset_index(), on='group'))

# переименуем колонку transactionId на orders
pivot_avg_orders.rename(columns={'transactionId': 'orders'}, inplace=True)

# создаем колонку со среднем количеством заказов на одного пользователя
pivot_avg_orders['avg_orders'] = pivot_avg_orders.orders.div(pivot_avg_orders.visitors)
# выводим таблицу на экран
pivot_avg_orders
```

Уже по таблице видна разница между средним количеством заказов на пользователя в разрезе групп, но для 100% уверенности проведем тест.

```python
# определяем критический уровень статистической значимости как 5%
alpha = 0.05 

# создаем массивы с покупками и пользователями для дальнейшего подсчета
successes = np.array(pivot_avg_orders.orders)
trials = np.array(pivot_avg_orders.visitors)

# пропорция успехов в первой группе:
p1 = successes[0]/trials[0]

# пропорция успехов во второй группе:
p2 = successes[1]/trials[1]

# пропорция успехов в комбинированном датасете:
p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])

# разница пропорций в датасетах
difference = p1 - p2 

# считаем статистику в ст.отклонениях стандартного нормального распределения
z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))

# задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
distr = st.norm(0, 1)  

p_value = (1 - distr.cdf(abs(z_value))) * 2

print('p-value: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: среднее количество заказов на пользователя из группы А отличается от среднего количества заказов на пользователя из группы В')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать среднее количество заказов на пользователя в разрезе групп разными'
    ) 
```

Мы убедились что среднее количество заказов отличается по группам.

3. Проанализировать результаты:
    - Если p-value меньше выбранного уровня значимости (alpha = 0.05), то отвергаем нулевую гипотезу и делаем вывод, что средний чек из группы *a*  отличается от среднего чека из группы *b*.
    - Если p-value больше выбранного уровня значимости, то нулевая гипотеза не может быть отвергнута и нельзя сделать вывод об отличии среднего чека из группы *a* от среднего чека из группы *b*.

```python
# определяем критический уровень статистической значимости как 5%
alpha = 0.05

results = st.mannwhitneyu(orders[orders['group']=='a']['revenue'], 
                          orders[orders['group']=='b']['revenue'])

print('p-значение: ', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу: разница статистически значима")
else:
    print('Не получилось отвергнуть нулевую гипотезу, нет оснований считать средней чек в разрезе групп разными')
```

Результаты теста показали, что не существуют статистически значимых различий в среднем чеке между двумя группами пользователей. Это означает, что на основе имеющихся данных, можно сделать вывод о том, что средний чек от пользователей между группами не отличается.
