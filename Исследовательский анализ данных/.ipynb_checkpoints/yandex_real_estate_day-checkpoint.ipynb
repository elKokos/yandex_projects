#### Работа с пропусками, типами данных, поиск дубликатов (явных и неявных) и аномалий, удаление выбросов в каждом столбце

создадим функцию для обнаружения, удаления выбросов и вывода графиков (коробки с усами), также функция будет возвращать обновленную таблицу

##### функции для предобработки данных

```python
# функция, использующая межквартильный диапазон
def remove_outliners_IQR(data_frame, column):
    # создаем копию датафрейма
    data_frame_after_remove = data_frame.copy()
    # вычисляем статистические характеристики столбца (квартили)
    Q1 = data_frame_after_remove[column].quantile(0.25)
    Q3 = data_frame_after_remove[column].quantile(0.75)
    IQR = Q3 - Q1
    # вычисляем границы межквартильного диапазона
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # удаляем выбросы из столбца
    data_frame_after_remove = data_frame_after_remove[(data_frame_after_remove[column] >= lower_bound) & (data_frame_after_remove[column] <= upper_bound)]
    
    # визуализируем графики до и после
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.boxplot(data_frame[column])
    plt.title('До удаления выбросов')
    plt.ticklabel_format(style='plain', axis='y')
    
    plt.subplot(1, 2, 2)
    plt.boxplot(data_frame_after_remove[column].dropna())
    plt.title('После удаления выбросов')
    plt.ticklabel_format(style='plain', axis='y')
    #plt.show()
    
    return data_frame_after_remove
```

```python
# создаем функцию, которая сформирует нам таблицу с количеством пропусков
def create_missing_df(data):
    # формируем дф с количеством пропусков в столбцах
    missing_df = data.isna().sum(axis=0).reset_index()
    missing_df.columns = ['column_name', 'missing_count']
    missing_df = missing_df[missing_df['missing_count'] > 0]
    missing_df = missing_df.sort_values('missing_count', ascending=True)
    # считаем % пропусков от общего
    missing_df['%'] = np.round((missing_df['missing_count'] / data.shape[0])*100, 2)
    return missing_df
```

```python
missing_df = create_missing_df(df)
missing_df.sort_values('%')
```

##### total_area, living_area и kitchen_area

living_area - 1903 (8.03%) и kitchen_area - 2278 (9.61%) пропущенных значений

<p>процент количества пропусков в этих столбцах превышает 5%, поэтому их удаление может повлиять на результаты анализа, а именно потерю значимых данных. Заполним пустые значения медианой</p>
<p>найдем медианное значение отношения жилой площади ко всей площади квартиры и округлим до одной цифры после запятой, тоже самое сделаем с площадью кухни</p>

```python
# найдем медианное значение отношения жилой площади ко всей площади квартиры и округлим до одной цифры после запятой, 
# тоже самое сделаем с площадью кухни
living_area_median = (df['living_area'] / df['total_area']).median().round(1)
kitchen_area_median = (df['kitchen_area'] / df['total_area']).median().round(1)
f"медианное значение отношения жилой площади ко всей площади квартиры - {living_area_median} и медианное значение отношения площади кухни ко всей площади квартиры {kitchen_area_median}"
```

```python
# создаем новые столбцы с новыми данными, в которых будут медианные значения жилой площади и кухни
df['median_living_area'] = df['total_area'] * living_area_median
df['median_kitchen_area'] = df['total_area'] *  kitchen_area_median
```

```python
# заменяем пустые значения на медианные
df['living_area'].fillna(df['median_living_area'], inplace=True)
df['kitchen_area'].fillna(df['median_kitchen_area'], inplace=True)
```

```python
# проверяем на наличие аномалий после заполнения пропусков
df[(df['living_area'] + df['kitchen_area']) >= df['total_area']].sort_values('total_area', ascending=False)
```

в ходе заполнения пропусков появились аномальные значения, так как не может быть такого, чтоб сумма жилой площади и площади кухни была выше общей площади квартиры. Аномалий менее 1% поэтому можно избавится от них, либо  заменить значения из новых столбцов с медианами. Выбираю второе. 

```python
df.loc[((df['living_area'] + df['kitchen_area']) >= df['total_area']).index, ['living_area']] = df['median_living_area']
df.loc[((df['living_area'] + df['kitchen_area']) >= df['total_area']).index, ['kitchen_area']] = df['median_kitchen_area']
```

#### функции для анализа данных

```python
def create_random_color():
    """
    данная функция генерирует случайный цвет 
    """
    r = random.random()
    g = random.random()
    b = random.random()
    return (r, g, b)
```

```python
def create_hist(column_name, data, bins=10, 
                xlabel='', ylabel='', title='', 
                color=None, alpha=0.7, grid=True, 
                ticklabel_format=None, figsize=(10, 5)):
    """
    Создает гистрограмму на основе указанного столбца из DataFrame.
    
    Параметры:
    - column_name: наименование столбца данных.
    - data: исходные данные.
    - bins: количество корзин (по умолчанию 10).
    - xlabel: подпись оси X (по умолчанию пустая строка).
    - ylabel: подпись оси Y (по умолчанию пустая строка).
    - title: заголовок гистограммы (по умолчанию пустая строка).
    - color: цвет столбцов гистограммы (по умолчанию None).
    - alpha: прозрачность столбцов гистограммы (по умолчанию 0.7).
    - grid: использовать сетку в гистограмме (по умолчанию True).
    - ticklabel_format: формат отображения значений на оси (по умолчанию None).
    - figsize: размер гистограммы (по умолчанию (10, 5)).
    """
    
    # если цвет не был определен при вызове функции, то создаем случайный цвет
    # иначе присваиваем цвет при вызове
    if color == None:
        color = create_random_color()
    
    # настраиваем гистограмму
    fig, ax = plt.subplots(figsize=figsize)
    ax.hist(data[column_name],
            bins=bins, 
            color=color, 
            alpha=alpha)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(grid)
    if ticklabel_format != None:
        try:
            plt.ticklabel_format(style=ticklabel_format)
        except:
            print("Try another format!")
```

```python
def create_scatter_plot(column_name_x, column_name_y,
                        data, xlabel='', ylabel='', 
                        title='', color=None, size=10, grid=True, 
                        figsize=(15, 5), alpha=0.1, ticklabel_format=None):
    """
    Создает диаграмму рассеивания на основе указанных столбцов из DataFrame.

    Параметры:
    - column_name_x: имя столбца, данные которого будут отображены на оси x.
    - column_name_y: имя столбца, данные которого будут отображены на оси y.
    - data: исходные данные.
    - xlabel: подпись оси X (по умолчанию пустая строка).
    - ylabel: подпись оси Y (по умолчанию пустая строка).
    - title: заголовок диаграммы (по умолчанию пустая строка).
    - color: цвет точек на диаграмме рассеивания (по умолчанию None).
    - size: размер точек на диаграмме рассеивания (по умолчанию 10).
    - grid: использовать сетку на диаграмме рассеивания (по умолчанию True).
    - figsize: размер диаграммы рассеивания (по умолчанию (15, 5)).
    - alpha: прозрачность точек на диаграмме рассеивания (по умолчанию 0.1).
    - ticklabel_format: формат отображения значений на оси (по умолчанию None).
    """
    if color == None:
        color = create_random_color()
    
    fig, ax = plt.subplots(figsize=figsize)
    ax.scatter(data[column_name_x], data[column_name_y], 
                color=color, s=size, alpha=alpha)
    
    plt.xlabel(xlabel if xlabel else column_name_x)
    plt.ylabel(ylabel if ylabel else column_name_y)
    plt.grid(grid)
    plt.title(title if title else 'диаграмма рассеивания')
    if ticklabel_format != None:
        try:
            plt.ticklabel_format(style=ticklabel_format)
        except:
            print("Try another format!")
```

```python
def create_corr_pearson(column_name):
    """
    персональная функция для третьего задания
    считает корреляцию Пирсона и возвращает результат
    """
    return df['last_price'].corr(df[column_name])
```
