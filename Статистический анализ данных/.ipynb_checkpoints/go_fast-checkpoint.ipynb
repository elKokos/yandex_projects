##### create_missing_df

```python
def create_missing_df(data):
    """
    —Ñ—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è —Å—Ñ–æ—Ä–º–∏—Ä—É–µ—Ç –Ω–∞–º —Ç–∞–±–ª–∏—Ü—É —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∏ –≤–µ—Ä–Ω–µ—Ç –µ–µ
    –Ω–∞ –≤—Ö–æ–¥ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω –ø–∞—Ä–∞–º–µ—Ç—Ä - DataFrame
    """
    # —Ñ–æ—Ä–º–∏—Ä—É–µ–º –¥—Ñ —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö
    missing_df = data.isna().sum(axis=0).reset_index()
    missing_df.columns = ['column_name', 'missing_count']
    missing_df = missing_df[missing_df['missing_count'] > 0]
    missing_df = missing_df.sort_values('missing_count', ascending=True)
    # —Å—á–∏—Ç–∞–µ–º % –ø—Ä–æ–ø—É—Å–∫–æ–≤ –æ—Ç –æ–±—â–µ–≥–æ
    missing_df['%'] = np.round((missing_df['missing_count'] / data.shape[0])*100, 2)
    return missing_df
```

##### remove_outliners_IQR

```python
def remove_outliners_IQR(data, column_name):
    """
    –§—É–Ω–∫—Ü–∏—è –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –º–µ–∂–∫–≤–∞—Ä—Ç–∏–ª—å–Ω—ã–π –¥–∏–∞–ø–∞–∑–æ–Ω –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è –≤—ã–±—Ä–æ—Å–æ–≤ –∏–∑ DataFrame,
    —Å–æ–∑–¥–∞–µ—Ç –≥—Ä–∞—Ñ–∏–∫–∏ –¥–æ –∏ –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –≤—ã–±—Ä–æ—Å–æ–≤,
    –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –æ–±–Ω–æ–≤–ª–µ–Ω–Ω—ã–π DataFrame –∏ DataFrame —Å –∞–Ω–æ–º–∞–ª–∏—è–º–∏.
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - data: –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ.
    - column_name: –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ –¥–∞–Ω–Ω—ã—Ö.
    """
    # —Å–æ–∑–¥–∞–µ–º –∫–æ–ø–∏—é —Ç–∞–±–ª–∏—Ü—ã
    data_after_remove = data.copy()
    # –≤—ã—á–∏—Å–ª—è–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ —Å—Ç–æ–ª–±—Ü–∞ (–∫–≤–∞—Ä—Ç–∏–ª–∏)
    Q1 = data_after_remove[column_name].quantile(0.25)
    Q3 = data_after_remove[column_name].quantile(0.75)
    IQR = Q3 - Q1
    # –≤—ã—á–∏—Å–ª—è–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –º–µ–∂–∫–≤–∞—Ä—Ç–∏–ª—å–Ω–æ–≥–æ –¥–∏–∞–ø–∞–∑–æ–Ω–∞
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # —É–¥–∞–ª—è–µ–º –≤—ã–±—Ä–æ—Å—ã –∏–∑ —Å—Ç–æ–ª–±—Ü–∞
    # –∑–∞–ø–∏—à–µ–º —É–¥–∞–ª–µ–Ω–Ω—ã–µ —Å—Ç—Ä–æ—á–∫–∏ –≤ –Ω–æ–≤—É—é —Ç–∞–±–ª–∏—Ü—É
    removed_df = data_after_remove[(data_after_remove[column_name] < lower_bound) | 
                                   (data_after_remove[column_name] > upper_bound)].copy()
    
    # –ø–µ—Ä–µ–∑–∞–ø–∏—à–µ–º –æ—Å–Ω–æ–≤–Ω—É—é —Ç–∞–±–ª–∏—Ü—É
    data_after_remove = data_after_remove[(data_after_remove[column_name] >= lower_bound) & 
                                          (data_after_remove[column_name] <= upper_bound)]
    
    # –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –≥—Ä–∞—Ñ–∏–∫–∏ –¥–æ –∏ –ø–æ—Å–ª–µ
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.boxplot(data[column_name])
    plt.title('–î–æ —É–¥–∞–ª–µ–Ω–∏—è –≤—ã–±—Ä–æ—Å–æ–≤')
    plt.ticklabel_format(style='plain', axis='y')
    
    plt.subplot(1, 2, 2)
    plt.boxplot(data_after_remove[column_name].dropna())
    plt.title('–ü–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –≤—ã–±—Ä–æ—Å–æ–≤')
    plt.ticklabel_format(style='plain', axis='y')
    
    return data_after_remove, removed_df
```

```python
def create_random_color():
    """
    –§—É–Ω–∫—Ü–∏—è –≥–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç —Å–ª—É—á–∞–π–Ω—ã–π —Ü–≤–µ—Ç 
    """
    r = random.random()
    g = random.random()
    b = random.random()
    return (r, g, b)
```

##### create_hist

```python
def create_hist(column_name, data, bins=10, 
                xlabel='', ylabel='', title='', 
                color=None, alpha=0.7, grid=True, 
                ticklabel_format=None, figsize=(10, 5)):
    """
    –§—É–Ω–∫—Ü–∏—è, —Å–æ–∑–¥–∞–µ—Ç –≥–∏—Å—Ç—Ä–æ–≥—Ä–∞–º–º—É –Ω–∞ –æ—Å–Ω–æ–≤–µ —É–∫–∞–∑–∞–Ω–Ω–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ –∏–∑ DataFrame.
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - column_name: –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ –¥–∞–Ω–Ω—ã—Ö.
    - data: –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ.
    - bins: –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ—Ä–∑–∏–Ω (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 10).
    - xlabel: –ø–æ–¥–ø–∏—Å—å –æ—Å–∏ X (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞).
    - ylabel: –ø–æ–¥–ø–∏—Å—å –æ—Å–∏ Y (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞).
    - title: –∑–∞–≥–æ–ª–æ–≤–æ–∫ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞).
    - color: —Ü–≤–µ—Ç —Å—Ç–æ–ª–±—Ü–æ–≤ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é None).
    - alpha: –ø—Ä–æ–∑—Ä–∞—á–Ω–æ—Å—Ç—å —Å—Ç–æ–ª–±—Ü–æ–≤ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 0.7).
    - grid: –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–µ—Ç–∫—É –≤ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–µ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é True).
    - ticklabel_format: —Ñ–æ—Ä–º–∞—Ç –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏–π –Ω–∞ –æ—Å–∏ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é None).
    - figsize: —Ä–∞–∑–º–µ—Ä –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é (10, 5)).
    """
    
    # –µ—Å–ª–∏ —Ü–≤–µ—Ç –Ω–µ –±—ã–ª –æ–ø—Ä–µ–¥–µ–ª–µ–Ω –ø—Ä–∏ –≤—ã–∑–æ–≤–µ —Ñ—É–Ω–∫—Ü–∏–∏, —Ç–æ —Å–æ–∑–¥–∞–µ–º —Å–ª—É—á–∞–π–Ω—ã–π —Ü–≤–µ—Ç
    # –∏–Ω–∞—á–µ –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ–º —Ü–≤–µ—Ç –ø—Ä–∏ –≤—ã–∑–æ–≤–µ
    if color == None:
        color = create_random_color()
    
    # –Ω–∞—Å—Ç—Ä–∞–∏–≤–∞–µ–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É
    fig, ax = plt.subplots(figsize=figsize)
    ax.hist(data[column_name],
            bins=bins, 
            color=color, 
            alpha=alpha)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(grid)
    if ticklabel_format != None:
        try:
            plt.ticklabel_format(style=ticklabel_format)
        except:
            print("Try another format!")
```

##### create_scatter_plot

```python
def create_scatter_plot(column_name_x, column_name_y,
                        data, xlabel='', ylabel='', 
                        title='', color=None, size=10, grid=True, 
                        figsize=(15, 5), alpha=0.07, ticklabel_format=None, trendline=False):
    """
    –°–æ–∑–¥–∞–µ—Ç –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è –Ω–∞ –æ—Å–Ω–æ–≤–µ —É–∫–∞–∑–∞–Ω–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –∏–∑ DataFrame.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - column_name_x: –∏–º—è —Å—Ç–æ–ª–±—Ü–∞, –¥–∞–Ω–Ω—ã–µ –∫–æ—Ç–æ—Ä–æ–≥–æ –±—É–¥—É—Ç –æ—Ç–æ–±—Ä–∞–∂–µ–Ω—ã –Ω–∞ –æ—Å–∏ x.
    - column_name_y: –∏–º—è —Å—Ç–æ–ª–±—Ü–∞, –¥–∞–Ω–Ω—ã–µ –∫–æ—Ç–æ—Ä–æ–≥–æ –±—É–¥—É—Ç –æ—Ç–æ–±—Ä–∞–∂–µ–Ω—ã –Ω–∞ –æ—Å–∏ y.
    - data: –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ.
    - xlabel: –ø–æ–¥–ø–∏—Å—å –æ—Å–∏ X (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞).
    - ylabel: –ø–æ–¥–ø–∏—Å—å –æ—Å–∏ Y (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞).
    - title: –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–∏–∞–≥—Ä–∞–º–º—ã (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞).
    - color: —Ü–≤–µ—Ç —Ç–æ—á–µ–∫ –Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º–µ —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é None).
    - size: —Ä–∞–∑–º–µ—Ä —Ç–æ—á–µ–∫ –Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º–µ —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 10).
    - grid: –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–µ—Ç–∫—É –Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º–µ —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é True).
    - figsize: —Ä–∞–∑–º–µ—Ä –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é (15, 5)).
    - alpha: –ø—Ä–æ–∑—Ä–∞—á–Ω–æ—Å—Ç—å —Ç–æ—á–µ–∫ –Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º–µ —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 0.07).
    - ticklabel_format: —Ñ–æ—Ä–º–∞—Ç –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏–π –Ω–∞ –æ—Å–∏ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é None).
    - trendline: –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –ª–∏–Ω–∏–∏ —Ç—Ä–µ–Ω–¥–∞ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é False).
    """
    
    if color == None:
        color = create_random_color()
    
    fig, ax = plt.subplots(figsize=figsize)
    ax.scatter(data[column_name_x], data[column_name_y], 
                color=color, s=size, alpha=alpha)
    
    plt.xlabel(xlabel if xlabel else column_name_x)
    plt.ylabel(ylabel if ylabel else column_name_y)
    plt.grid(grid)
    plt.title(title if title else '–¥–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è')
    if ticklabel_format != None:
        try:
            plt.ticklabel_format(style=ticklabel_format)
        except:
            print("Try another format!")
            
    if trendline:
        slope, intercept = np.polyfit(data[column_name_x], data[column_name_y], 1)
        trendline = slope * np.array(data[column_name_x]) + intercept
        plt.plot(data[column_name_x], trendline, color=create_random_color())
```

##### create_pie_chart

```python
def create_pie_chart(data, column_name, title='', figsize=(15, 10), shadow=True):
    """
    –°–æ–∑–¥–∞–µ—Ç –∫—Ä—É–≥–æ–≤—É—é –¥–∏–∞–≥—Ä–∞–º–º—É –Ω–∞ –æ—Å–Ω–æ–≤–µ —É–∫–∞–∑–∞–Ω–Ω–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ –∏–∑ DataFrame.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - data: –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ.
    - column_name: –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ –¥–∞–Ω–Ω—ã—Ö.
    - title: –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–∏–∞–≥—Ä–∞–º–º—ã (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞).
    - figsize: —Ä–∞–∑–º–µ—Ä –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é (15, 10)).
    - shadow: –Ω–∞–ª–∏—á–∏–µ —Ç–µ–Ω–∏ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é True).
    """
    
    # —Å–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ —Ü–≤–µ—Ç–æ–≤ –¥–ª—è –±—É–¥—É—â–µ–π –¥–∏–∞–≥—Ä–∞–º–º—ã
    # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–≤–µ—Ç–æ–≤ = –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ
    colors = []
    for _ in range(data[column_name].nunique()):
        colors.append(create_random_color())
    # —Å–æ–∑–¥–∞–µ–º —Ä–∏—Å—É–Ω–æ–∫
    fig, ax = plt.subplots(figsize=figsize)
    # –Ω–∞—Å—Ç—Ä–∞–∏–≤–∞–µ–º –¥–∏–∞–≥—Ä–∞–º–º—É
    # –≤ –∫–∞—á–µ—Å—Ç–≤–µ –¥–∞–Ω–Ω—ã—Ö –∏—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥ value_counts + —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ sort_values
    # –∞ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ª—ç–π–±–ª–æ–≤ –∏—Å–ø–æ–ª—å–∑—É–µ–º value_counts, sort_values –∏ –¥–æ—Å—Ç–∞–µ–º –∏–Ω–¥–µ–∫—Å—ã
    ax.pie(data[column_name].value_counts().sort_values(), autopct='%1.1f%%', 
           labels=data[column_name].value_counts().sort_values().index, 
           shadow=shadow, colors=colors, textprops={'fontsize': 14})
    
    ax.set_title(title if title else '–∫—Ä—É–≥–æ–≤–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞', fontsize=20)
```

##### –í—ã–≤–æ–¥ 
–ü—Ä–∏ –ø—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ–º –∑–Ω–∞–∫–æ–º—Å—Ç–≤–µ —Å –¥–∞–Ω–Ω—ã–º–∏ –æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è—Ö —Å–µ—Ä–≤–∏—Å–∞ –∞—Ä–µ–Ω–¥—ã —Å–∞–º–æ–∫–∞—Ç–æ–≤ –º–æ–∂–Ω–æ –æ—Ç–º–µ—Ç–∏—Ç—å —Å–ª–µ–¥—É—é—â–µ–µ:

##### –û–±—ä–µ–¥–∏–Ω–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ –æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è—Ö, –ø–æ–µ–∑–¥–∫–∞—Ö –∏ –ø–æ–¥–ø–∏—Å–∫–∞—Ö –≤ –æ–¥–∏–Ω –¥–∞—Ç–∞—Ñ—Ä–µ–π–º.
–î–ª—è —ç—Ç–æ–≥–æ –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å –º–µ—Ç–æ–¥–æ–º merge().

```python
# –æ–±—ä–µ–¥–µ–Ω–∏–º –≤—Å–µ —Ç—Ä–∏ –≤—ã–±–æ—Ä–∫–∏ –≤ –æ–¥–Ω—É –±–æ–ª—å—à—É—é —Ç–∞–±–ª–∏—Ü—É
# –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è –º–µ—Ç–æ–¥–æ–º merge –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ pandas
# —Å–¥–µ–ª–∞–µ–º  –≤—Å—ë —ç—Ç–æ –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ—á–∫—É
# –≤ –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ —Å–Ω–∞—á–∞–ª–∞ –æ–±—ä–µ–¥–∏–Ω—è–µ–º –≤—ã–±–æ—Ä–∫–∏ users_df –∏ rides_df –ø–æ –∫–æ–ª–æ–Ω–∫–µ 'user_id'
# –∞ –∑–∞—Ç–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç —ç—Ç–æ–≥–æ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è —Å–æ–µ–¥–∏–Ω—è–µ–º  —Å –≤—ã–±–æ—Ä–∫–æ–π subs_df –ø–æ –∫–æ–ª–æ–Ω–∫–µ 'subscription_type'
df = pd.merge(pd.merge(users_df, rides_df, on='user_id'), subs_df, on='subscription_type')
```

```python
# –ø—Ä–æ–≤–µ—Ä—è–µ–º –ø–æ–ª—É—á–∏–ª–æ—Å—å –ª–∏ —Å–æ–µ–¥–∏–Ω–∏—Ç—å –≤—ã–±–æ—Ä–∫–∏
# –∏ –∑–∞ –æ–¥–Ω–æ –ø—Ä–æ–≤–µ—Ä–∏–º –Ω–æ–≤—É—é —Ç–∞–±–ª–∏—Ü—É –Ω–∞ –Ω–∞–ª–∏—á–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∏–ª–∏ –µ—â–µ —á–µ–≥–æ-–ª–∏–±–æ –Ω–µ–ø—Ä–µ–¥–≤–∏–¥–µ–Ω–Ω–æ–≥–æ
describe_data(df)
```

##### –°–æ–∑–¥–∞–π—Ç–µ –µ—â—ë –¥–≤–∞ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞, —Å–æ–∑–¥–∞–Ω–Ω–æ–≥–æ –Ω–∞ —ç—Ç–∞–ø–µ 4.1

- c –¥–∞–Ω–Ω—ã–º–∏ –æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è—Ö –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏.

```python
# —Å–æ–∑–¥–∞–¥–∏–º —Ç–∞–±–ª–∏—Ü—É —Å –¥–∞–Ω–Ω—ã–º–∏ –æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è—Ö —Å –ø–æ–¥–ø–∏—Å–∫–æ–π (–±–µ–∑ –∫–æ–ø–∏—Ä–æ–≤–∞–Ω–∏—è)
ultra_df = df.loc[df['subscription_type'] == 'ultra']
# –≤—ã–≤–µ–¥–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –Ω–æ–≤–æ–π —Ç–∞–±–ª–∏—Ü–µ
describe_data(ultra_df)
```

```python
# —Å–æ–∑–¥–∞–¥–∏–º —Ç–∞–±–ª–∏—Ü—É —Å –¥–∞–Ω–Ω—ã–º–∏ –æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è—Ö –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏ (–±–µ–∑ –∫–æ–ø–∏—Ä–æ–≤–∞–Ω–∏—è)
free_df = df.loc[df['subscription_type'] == 'free']
# –≤—ã–≤–µ–¥–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –Ω–æ–≤–æ–π —Ç–∞–±–ª–∏—Ü–µ
describe_data(free_df)
```

#### –®–∞–≥ 5. –ü–æ–¥—Å—á—ë—Ç –≤—ã—Ä—É—á–∫–∏

##### –°–æ–∑–¥–∞–π—Ç–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –æ –ø–æ–µ–∑–¥–∫–∞—Ö –Ω–∞ –æ—Å–Ω–æ–≤–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ —Å –æ–±—ä–µ–¥–∏–Ω—ë–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –∏–∑ —à–∞–≥–∞ 4:
–Ω–∞–π–¥–∏—Ç–µ —Å—É–º–º–∞—Ä–Ω–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–µ–∑–¥–æ–∫ –∏ —Å—É–º–º–∞—Ä–Ω–æ–µ –≤—Ä–µ–º—è –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∑–∞ –∫–∞–∂–¥—ã–π –º–µ—Å—è—Ü.

–ß—Ç–æ–±—ã –≤—ã–ø–æ–ª–Ω–∏—Ç—å —ç—Ç—É –∑–∞–¥–∞—á—É –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ: 
- —Å–æ–∑–¥–∞—Ç—å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π —Å—Ç–æ–ª–±–∏–∫ —Å–æ –∑–Ω–∞—á–µ–Ω–∏–µ–º –º–µ—Å—è—Ü–∞ –≤ df
- —Å–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ –ø–æ  –º–µ—Å—è—Ü—É –∏ id –∏ –¥–æ–±–∞–≤–∏—Ç—å –∞–≥–≥—Ä–µ–≥–∞—Ü–∏—é –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–º —Å—Ç–æ–ª–±—Ü–∞–º (aggregated_df)
- —Å–±—Ä–æ—Å–∏–º –∏–Ω–¥–µ–∫—Å—ã –∏ –¥–æ–±–∞–≤–∏–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ —Å—Ç–æ–ª–±—Ü—ã –∏–∑ –æ—Å–Ω–æ–≤–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã (df) –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –ø–æ—Å–ª–µ–¥—É—é—â–∏—Ö –∑–∞–¥–∞—á (–∞ –∏–º–µ–Ω–Ω–æ –ø–æ–¥—Å—á–µ—Ç–∞ –≤—ã—Ä—É—á–∫–∏)

```python
df['month'] = df.date.dt.month
```

```python
aggregated_df = df.groupby(['user_id', 'month']).agg({
    'duration' : 'sum',
    'distance' : 'sum',
    'month' : 'count',
})
aggregated_df.rename(columns={'month' : 'count'}, inplace=True)
```

```python
aggregated_df.reset_index(inplace=True, level=['month', 'user_id'])
aggregated_df.head(20)
```

```python
aggregated_df = (aggregated_df.merge(df[['user_id','minute_price', 'start_ride_price', 
                                         'subscription_fee', 'subscription_type']], 
                                    on='user_id'))
```

##### –í–∞–∂–Ω–æ –ø–æ–Ω—è—Ç—å, —Ç—Ä–∞—Ç—è—Ç –ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –±–æ–ª—å—à–µ –≤—Ä–µ–º–µ–Ω–∏ –Ω–∞ –ø–æ–µ–∑–¥–∫–∏?

1. –°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –≥–∏–ø–æ—Ç–µ–∑—É:
    - –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H0): –°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –ø–æ–µ–∑–¥–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π —Ä–∞–≤–Ω–æ —Å—Ä–µ–¥–Ω–µ–º—É –≤—Ä–µ–º–µ–Ω–∏ –ø–æ–µ–∑–¥–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏.
    - –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H1): –°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –ø–æ–µ–∑–¥–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –±–æ–ª—å—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ–µ–∑–¥–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏.

2. –í—ã–ø–æ–ª–Ω–∏—Ç—å —Ç–µ—Å—Ç:
    - –í—ã–ø–æ–ª–Ω—è–µ–º t-—Ç–µ—Å—Ç, –∏—Å–ø–æ–ª—å–∑—É—è —Ñ—É–Ω–∫—Ü–∏—é ttest_ind(). –£—Å—Ç–∞–Ω–æ–≤–ª–∏–≤–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä equal_var –≤ False, —á—Ç–æ–±—ã —É–∫–∞–∑–∞—Ç—å, —á—Ç–æ –º—ã —Å—á–∏—Ç–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—ã–±–æ—Ä–æ–∫ —Ä–∞–∑–Ω—ã–º–∏.
    - –ü–æ–ª—É—á–∏—Ç—å p-value —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ .pvalue.
    
3. –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã:
    - –ï—Å–ª–∏ p-value –º–µ–Ω—å—à–µ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ (alpha = 0.05), —Ç–æ –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –∏ –¥–µ–ª–∞–µ–º –≤—ã–≤–æ–¥, —á—Ç–æ —Å—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –ø–æ–µ–∑–¥–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ–µ–∑–¥–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏.
    - –ï—Å–ª–∏ p-value –±–æ–ª—å—à–µ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏, —Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç–∞ –∏ –Ω–µ–ª—å–∑—è —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥ –æ–± –æ—Ç–ª–∏—á–∏–∏ —Å—Ä–µ–¥–Ω–µ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ–µ–∑–¥–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –æ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ–µ–∑–¥–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏.


```python
# —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 5%
alpha = .05
# –∑–∞–ø–∏—à–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –º–µ—Ç–æ–¥–∞ ttest_ind –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é results_duration,
# –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø–µ—Ä–≤–æ–≥–æ –ø–∞—Ä–∞–º–µ–Ω—Ç—Ä–∞ —É–∫–∞–∂–µ–º –≤—ã–±–æ—Ä–∫—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π, –∞ –≤—Ç–æ—Ä–æ–≥–æ - –≤—ã–±–æ—Ä–∫–∞ –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏
# —É–∫–∞–∂–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä equal_var - False, 
# —Ç–∞–∫ –∫–∞–∫ –≤—ã–±–æ—Ä–∫–∏ —É –Ω–∞—Å —Ä–∞–∑–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ –∏ —Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —É –≤—ã–±–æ—Ä–æ–∫ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è
# —É–∫–∞–∂–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä alternative - greater –¥–ª—è –ø—Ä–∞–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ–π –≥–∏–ø–æ—Ç–µ–∑—ã
results_duration = st.ttest_ind(ultra_df['duration'], free_df['duration'], 
                                equal_var=False, alternative='greater')
# –æ–∑–Ω–∞–∫–æ–º–∏–º—Å—è —Å pvalue
print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results_duration.pvalue)
# —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–π pvalue —Å —É—Ä–æ–≤–Ω–µ–º –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ (alpha)
# –µ—Å–ª–∏ pvalue –º–µ–Ω—å—à–µ alpha, —Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–∫–ª–æ–Ω—è–µ—Ç—Å—è
# –∏–Ω–∞—á–µ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç—Å—è
if results_duration.pvalue < alpha:
    print('–†–∞–∑–Ω–∏—Ü–∞ –≤–æ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ–µ–∑–¥–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –∏ –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞')
else:
    print('–ù–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ–π —Ä–∞–∑–Ω–∏—Ü—ã –≤–æ –≤—Ä–µ–º–µ–Ω–∏')
```

–ù–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤, –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥ –æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ —Ä–∞–∑–Ω–∏—Ü—ã –≤—Ä–µ–º–µ–Ω–∏ –ø–æ–µ–∑–¥–æ–∫ –Ω–∞ —Å–∞–º–æ–∫–∞—Ç–∞—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –∏ –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏. –í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è –º–µ—Ç–æ–¥–∞ ttest_ind –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ scipy –∫ –¥–≤—É–º –≤—ã–±–æ—Ä–∫–∞–º, –∑–Ω–∞—á–µ–Ω–∏—è p_value –ø–æ–ª—É—á–∏–ª–∏—Å—å –º–µ–Ω—å—à–µ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ (0.05). –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ —ç—Ç–∏—Ö –¥–∞–Ω–Ω—ã—Ö –º–æ–∂–µ–º —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥, —á—Ç–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –ø–æ–¥–ø–∏—Å–∫–æ–π —Ç—Ä–∞—Ç—è—Ç –±–æ–ª—å—à–µ –≤—Ä–µ–º–µ–Ω–∏ –Ω–∞ –ø–æ–µ–∑–¥–∫–∏ –Ω–∞ —Å–∞–º–æ–∫–∞—Ç–∞—Ö.

##### –†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –æ–¥–Ω–æ–π –ø–æ–µ–∑–¥–∫–∏ –≤ 3130 –º–µ—Ç—Ä–æ–≤ ‚Äî –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ —Å —Ç–æ—á–∫–∏ –∑—Ä–µ–Ω–∏—è –∏–∑–Ω–æ—Å–∞ —Å–∞–º–æ–∫–∞—Ç–∞. 
–ú–æ–∂–Ω–æ –ª–∏ —Å–∫–∞–∑–∞—Ç—å, —á—Ç–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –ø—Ä–æ–µ–∑–∂–∞—é—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –∑–∞ –æ–¥–Ω—É –ø–æ–µ–∑–¥–∫—É, –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç 3130 –º–µ—Ç—Ä–æ–≤? –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –≥–∏–ø–æ—Ç–µ–∑—É –∏ —Å–¥–µ–ª–∞–π—Ç–µ –≤—ã–≤–æ–¥—ã.

–î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –¥–∞–Ω–Ω–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã, –∏—Å–ø–æ–ª—å–∑—É–µ–º –æ–¥–Ω–æ–≤—ã–±–æ—Ä–æ—á–Ω—ã–π T-—Ç–µ—Å—Ç (ttest_1samp).
1. –°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –≥–∏–ø–æ—Ç–µ–∑—É:
    - –ü—É—Å—Ç—å –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H0): - —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –ø—Ä–æ–µ–∑–∂–∞—é—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –∑–∞ –æ–¥–Ω—É –ø–æ–µ–∑–¥–∫—É, –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç 3130 –º–µ—Ç—Ä–æ–≤.
    - –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H1): - —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –ø—Ä–æ–µ–∑–∂–∞—é—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –∑–∞ –æ–¥–Ω—É –ø–æ–µ–∑–¥–∫—É, –ø—Ä–µ–≤—ã—à–∞–µ—Ç 3130 –º–µ—Ç—Ä–æ–≤.
2. –í—ã–ø–æ–ª–Ω–∏—Ç—å —Ç–µ—Å—Ç:
    - –û–ø—Ä–µ–¥–µ–ª–∏–º –Ω–∞—à–µ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ. 
    - –í—ã–ø–æ–ª–Ω—è–µ–º t-—Ç–µ—Å—Ç, –∏—Å–ø–æ–ª—å–∑—É—è —Ñ—É–Ω–∫—Ü–∏—é ttest_1samp().
    - –ü–æ–ª—É—á–∏—Ç—å p-value —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ .pvalue.
    
3. –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã:
    - –ï—Å–ª–∏ p-value –º–µ–Ω—å—à–µ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ (alpha = 0.05), —Ç–æ –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –∏ –¥–µ–ª–∞–µ–º –≤—ã–≤–æ–¥, —á—Ç–æ —Å—Ä–µ–¥–Ω–µ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –ø—Ä–µ–æ–¥–æ–ª–µ–≤–∞–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —Å –ø–æ–¥–ø–∏—Å–∫–æ–π, –ø—Ä–µ–≤—ã—à–∞–µ—Ç 3130 –º–µ—Ç—Ä–æ–≤.
    - –ï—Å–ª–∏ p-value –±–æ–ª—å—à–µ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏, —Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç–∞ –∏ –Ω–µ–ª—å–∑—è —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥, —á—Ç–æ —Å—Ä–µ–¥–Ω–µ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –ø—Ä–µ–æ–¥–æ–ª–µ–≤–∞–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —Å –ø–æ–¥–ø–∏—Å–∫–æ–π, –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç 3130 –º–µ—Ç—Ä–æ–≤.

```python
# —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 5% –∏ –æ–Ω –æ–ø—Ä–µ–¥–µ–ª–µ–Ω –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π alpha
# —Å–æ—Ö—Ä–∞–Ω–∏–º –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é interested_value –Ω–∞—à–µ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
interested_value = 3130
# –∑–∞–ø–∏—à–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –º–µ—Ç–æ–¥–∞ ttest_1samp –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é results_distance,
# –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø–∞—Ä–∞–º–µ–Ω—Ç—Ä–∞ —É–∫–∞–∂–µ–º –≤—ã–±–æ—Ä–∫—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π
results_distance = st.ttest_1samp(ultra_df['distance'], interested_value, alternative='greater')
# –æ–∑–Ω–∞–∫–æ–º–∏–º—Å—è —Å pvalue
print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results_distance.pvalue)
# —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–π pvalue —Å —É—Ä–æ–≤–Ω–µ–º –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ (alpha)
# –µ—Å–ª–∏ pvalue –º–µ–Ω—å—à–µ alpha, —Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–∫–ª–æ–Ω—è–µ—Ç—Å—è
# –∏–Ω–∞—á–µ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç—Å—è
if results_distance.pvalue < alpha:
    print('–û—Ç–∫–ª–æ–Ω—è–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: –°—Ä–µ–¥–Ω–µ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç 3130 –º–µ—Ç—Ä–æ–≤')
else:
    print('–ù–µ –æ—Ç–∫–ª–æ–Ω—è–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: –°—Ä–µ–¥–Ω–µ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç 3130 –º–µ—Ç—Ä–æ–≤')
```

–ò—Ç–∞–∫, –ø–æ–ª—É—á–µ–Ω–Ω–æ–µ p-value —Ä–∞–≤–Ω–æ 0.0317, —á—Ç–æ –º–µ–Ω—å—à–µ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏. –°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, —É –Ω–∞—Å –µ—Å—Ç—å –æ—Å–Ω–æ–≤–∞–Ω–∏—è –æ—Ç–≤–µ—Ä–≥–∞—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É. –ú—ã –º–æ–∂–µ–º —Å–∫–∞–∑–∞—Ç—å, —á—Ç–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –ø—Ä–æ–µ–∑–∂–∞—é—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –∑–∞ –æ–¥–Ω—É –ø–æ–µ–∑–¥–∫—É, –ø—Ä–µ–≤—ã—à–∞–µ—Ç 3130 –º–µ—Ç—Ä–æ–≤.

##### –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, –±—É–¥–µ—Ç –ª–∏ –ø–æ–º–µ—Å—è—á–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –ø–æ –º–µ—Å—è—Ü–∞–º –≤—ã—à–µ, —á–µ–º –≤—ã—Ä—É—á–∫–∞ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏. 

–î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –¥–∞–Ω–Ω–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã, –∏—Å–ø–æ–ª—å–∑—É–µ–º –¥–ª—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö –≤—ã–±–æ—Ä–æ–∫ (ttest_ind).
1. –°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –≥–∏–ø–æ—Ç–µ–∑—É:
    - –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H0): –í—ã—Ä—É—á–∫–∏ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –∏ –±–µ–∑ –Ω–µ—ë —Ä–∞–≤–Ω–æ–∑–Ω–∞—á–Ω—ã.
    - –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H1): –í—ã—Ä—É—á–∫–∞ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –ø–æ –º–µ—Å—è—Ü–∞–º –≤—ã—à–µ, —á–µ–º –≤—ã—Ä—É—á–∫–∞ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏.

2. –í—ã–ø–æ–ª–Ω–∏—Ç—å —Ç–µ—Å—Ç:
    - –í—ã–ø–æ–ª–Ω—è–µ–º t-—Ç–µ—Å—Ç, –∏—Å–ø–æ–ª—å–∑—É—è —Ñ—É–Ω–∫—Ü–∏—é ttest_ind(). –£—Å—Ç–∞–Ω–æ–≤–ª–∏–≤–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä equal_var –≤ False, —á—Ç–æ–±—ã —É–∫–∞–∑–∞—Ç—å, —á—Ç–æ –º—ã —Å—á–∏—Ç–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—ã–±–æ—Ä–æ–∫ —Ä–∞–∑–Ω—ã–º–∏.
    - –ü–æ–ª—É—á–∏—Ç—å p-value —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ .pvalue.
    
3. –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã:
    - –ï—Å–ª–∏ p-value –º–µ–Ω—å—à–µ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ (alpha = 0.05), —Ç–æ –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –∏ –¥–µ–ª–∞–µ–º –≤—ã–≤–æ–¥, —á—Ç–æ –≤—ã—Ä—É—á–∫–∞ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –≤ —Å—Ä–µ–¥–Ω–µ–º –≤—ã—à–µ.
    - –ï—Å–ª–∏ p-value –±–æ–ª—å—à–µ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏, —Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç–∞ –∏ –Ω–µ–ª—å–∑—è —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥ –æ–± –æ—Ç–ª–∏—á–∏–∏ –≤—ã—Ä—É—á–∫–∏ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –∏ –±–µ–∑ –Ω–µ—ë.

```python
# —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 5% –∏ –æ–Ω –æ–ø—Ä–µ–¥–µ–ª–µ–Ω –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π alpha

# –∑–∞–ø–∏—à–µ–º –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é ultra_month_revenue –≤—ã—Ä—É—á–∫—É –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π
ultra_month_revenue = aggregated_df[aggregated_df['subscription_type'] == 'ultra']['revenue']

# –∑–∞–ø–∏—à–µ–º –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é free_month_revenue –≤—ã—Ä—É—á–∫—É –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏
free_month_revenue = aggregated_df[aggregated_df['subscription_type'] == 'free']['revenue']

# –∑–∞–ø–∏—à–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –º–µ—Ç–æ–¥–∞ ttest_ind –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é results_distance,
# –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø–∞—Ä–∞–º–µ–Ω—Ç—Ä–æ–≤ —É–∫–∞–∂–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ ultra_month_revenue –∏ free_month_revenue
# —É–∫–∞–∂–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä equal_var - False, 
# —Ç–∞–∫ –∫–∞–∫ –≤—ã–±–æ—Ä–∫–∏ —É –Ω–∞—Å —Ä–∞–∑–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ –∏ —Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —É –≤—ã–±–æ—Ä–æ–∫ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è
# —É–∫–∞–∂–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä alternative - greater –¥–ª—è –ø—Ä–∞–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ–π –≥–∏–ø–æ—Ç–µ–∑—ã
results_revenue = st.ttest_ind(ultra_month_revenue, free_month_revenue, 
                               equal_var=False, alternative='greater')

# –æ–∑–Ω–∞–∫–æ–º–∏–º—Å—è —Å pvalue
print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results_revenue.pvalue)
# —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–π pvalue —Å —É—Ä–æ–≤–Ω–µ–º –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ (alpha)
# –µ—Å–ª–∏ pvalue –º–µ–Ω—å—à–µ alpha, —Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–∫–ª–æ–Ω—è–µ—Ç—Å—è
# –∏–Ω–∞—á–µ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç—Å—è
if results_revenue.pvalue < alpha:
    print('–û—Ç–∫–ª–æ–Ω—è–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: –µ—Å—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–µ —Ä–∞–∑–ª–∏—á–∏—è –≤ –≤—ã—Ä—É—á–∫–µ –º–µ–∂–¥—É –¥–≤—É–º—è –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ –º–µ—Å—è—Ü–∞–º')
else:
    print('–ù–µ –æ—Ç–∫–ª–æ–Ω—è–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: —É –Ω–∞—Å –Ω–µ—Ç –¥–æ—Å—Ç–∞—Ç–æ—á–Ω—ã—Ö –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤ –¥–ª—è –æ—Ç–≤–µ—Ä–∂–µ–Ω–∏—è –Ω—É–ª–µ–≤–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ –≤—ã—Ä—É—á–∫–∏ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –∏ –±–µ–∑ –ø–æ–¥–ø–∏—Å–∫–∏.')
```
