##### describe_data

```python
def describe_data(data):
    """
     Выводит основные характеристики DataFrame, включая:
    - 10 случайных строк из таблицы
    - размерность DataFrame
    - информацию по числовым столбцам
    - информацию по строковым столбцам
    - количество пропусков в таблице
    - количество уникальных значений в каждом столбце
    - количество дублирующихся строк
    - количество уникальных значений в каждом столбце
    - гистограммы по всем столбцам
    """
    # вывод 10 случайных строк
    print('\033[92mвывод случайных строк из таблицы:\033[90m\n')
    # если строк в таблице меньше 10, то выскочит ошибка
    # отлавливаем ее и выводим все имеющиеся строки в таблице
    try:
        display(data.sample(10))
    except:
        display(data.sample(data.shape[0]))
    
    # вывод типов данных в столбцах
    print('Типы данных:')
    print(data.dtypes)
    ('-----------------------------------------------------------------------')
    
    # вывод количества строк
    print('\033[92mКоличество строк:\033[90m', f'\033[91m{data.shape[0]}\033[90m')
    print('-----------------------------------------------------------------------')
    
    # вывод количества столбцов 
    print('\033[92mКоличество столбцов:\033[90m', f'\033[91m{data.shape[1]}\033[90m')
    print('-----------------------------------------------------------------------')
    
    # вывод информации о числовых столбцах
    print('\033[92mИнформация по числовым столбцам:\033[90m\n')
    for column in data.select_dtypes(include='number'):
        print(data[column].describe(), '\n')
        print('-----------------------------------------------------------------------')
    
    # вывод информации об оъектных данных
    print('\033[92mИнформация по строковым столбцам:\033[90m\n')
    for column in data.select_dtypes(include='object'):
        print(data[column].describe(), '\n')
        print('-----------------------------------------------------------------------')
    
    # вывод столбцов, содержащих пропуски, и их количество
    print('\033[92mКоличество пропусков в таблице:\033[90m\n')
    # эта переменная нужно для подсчета столбиков с пропусками
    count_na = 0
    for column in data.columns:
        null_in_column = data[column].isna().sum()
        
        if null_in_column:
            count += 1
            print(f'\033[91m{column}: {null_in_column}\033[90m')
            print('-----------------------------------------------------------------------')
    if count_na == 0:
        print('В таблице нет пропусков 🥂')
    
    # вывод информации об уникальных значениях каждого столбца
    print('\033[92mКоличество уникальных значений в каждом столбце:\033[90m\n')
    for column in data.columns:
        unique_values = data[column].nunique()
        print(f'{column} : {unique_values}')
        print('-----------------------------------------------------------------------')
    
    # вывод информации о дублирующих строках
    print('\n')
    duplicated_value = data.duplicated().sum()
    if duplicated_value:
        print(f'Количество дублирующихся строк: \033[91m{duplicated_value}\033[90m')
    else:
        print('\033[94mДублирующихся строк не обнаружено\033[90m')
    print('\n')
    
    # построение гистограммы по всем столбцам (для ознакомления)
    print('\n')
    print('\033[92mГистограммы по всем столбцам:\033[90m\n')
    try:
        data.hist(figsize=(10, 10))
    except:
        print('не получилось построить гистограммы')
```

##### create_missing_df

```python
def create_missing_df(data):
    """
    функция, которая сформирует нам таблицу с количеством пропусков и вернет ее
    на вход принимает только один параметр - DataFrame
    """
    # формируем дф с количеством пропусков в столбцах
    missing_df = data.isna().sum(axis=0).reset_index()
    missing_df.columns = ['column_name', 'missing_count']
    missing_df = missing_df[missing_df['missing_count'] > 0]
    missing_df = missing_df.sort_values('missing_count', ascending=True)
    # считаем % пропусков от общего
    missing_df['%'] = np.round((missing_df['missing_count'] / data.shape[0])*100, 2)
    return missing_df
```

##### remove_outliners_IQR

```python
def remove_outliners_IQR(data, column_name):
    """
    Функция использует межквартильный диапазон для удаления выбросов из DataFrame,
    создает графики до и после удаления выбросов,
    возвращает обновленный DataFrame и DataFrame с аномалиями.
    Параметры:
    - data: исходные данные.
    - column_name: наименование столбца данных.
    """
    # создаем копию таблицы
    data_after_remove = data.copy()
    # вычисляем статистические характеристики столбца (квартили)
    Q1 = data_after_remove[column_name].quantile(0.25)
    Q3 = data_after_remove[column_name].quantile(0.75)
    IQR = Q3 - Q1
    # вычисляем границы межквартильного диапазона
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # удаляем выбросы из столбца
    # запишем удаленные строчки в новую таблицу
    removed_df = data_after_remove[(data_after_remove[column_name] < lower_bound) | 
                                   (data_after_remove[column_name] > upper_bound)].copy()
    
    # перезапишем основную таблицу
    data_after_remove = data_after_remove[(data_after_remove[column_name] >= lower_bound) & 
                                          (data_after_remove[column_name] <= upper_bound)]
    
    # визуализируем графики до и после
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.boxplot(data[column_name])
    plt.title('До удаления выбросов')
    plt.ticklabel_format(style='plain', axis='y')
    
    plt.subplot(1, 2, 2)
    plt.boxplot(data_after_remove[column_name].dropna())
    plt.title('После удаления выбросов')
    plt.ticklabel_format(style='plain', axis='y')
    
    return data_after_remove, removed_df
```

##### create_random_color

```python
def create_random_color():
    """
    Функция генерирует случайный цвет 
    """
    r = random.random()
    g = random.random()
    b = random.random()
    return (r, g, b)
```

##### create_hist

```python
def create_hist(column_name, data, bins=10, 
                xlabel='', ylabel='', title='', 
                color=None, alpha=0.7, grid=True, 
                ticklabel_format=None, figsize=(10, 5)):
    """
    Функция, создает гистрограмму на основе указанного столбца из DataFrame.
    
    Параметры:
    - column_name: наименование столбца данных.
    - data: исходные данные.
    - bins: количество корзин (по умолчанию 10).
    - xlabel: подпись оси X (по умолчанию пустая строка).
    - ylabel: подпись оси Y (по умолчанию пустая строка).
    - title: заголовок гистограммы (по умолчанию пустая строка).
    - color: цвет столбцов гистограммы (по умолчанию None).
    - alpha: прозрачность столбцов гистограммы (по умолчанию 0.7).
    - grid: использовать сетку в гистограмме (по умолчанию True).
    - ticklabel_format: формат отображения значений на оси (по умолчанию None).
    - figsize: размер гистограммы (по умолчанию (10, 5)).
    """
    
    # если цвет не был определен при вызове функции, то создаем случайный цвет
    # иначе присваиваем цвет при вызове
    if color == None:
        color = create_random_color()
    
    # настраиваем гистограмму
    fig, ax = plt.subplots(figsize=figsize)
    ax.hist(data[column_name],
            bins=bins, 
            color=color, 
            alpha=alpha)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(grid)
    if ticklabel_format != None:
        try:
            plt.ticklabel_format(style=ticklabel_format)
        except:
            print("Try another format!")
```

##### create_scatter_plot

```python
def create_scatter_plot(column_name_x, column_name_y,
                        data, xlabel='', ylabel='', 
                        title='', color=None, size=10, grid=True, 
                        figsize=(15, 5), alpha=0.07, ticklabel_format=None, trendline=False):
    """
    Создает диаграмму рассеивания на основе указанных столбцов из DataFrame.

    Параметры:
    - column_name_x: имя столбца, данные которого будут отображены на оси x.
    - column_name_y: имя столбца, данные которого будут отображены на оси y.
    - data: исходные данные.
    - xlabel: подпись оси X (по умолчанию пустая строка).
    - ylabel: подпись оси Y (по умолчанию пустая строка).
    - title: заголовок диаграммы (по умолчанию пустая строка).
    - color: цвет точек на диаграмме рассеивания (по умолчанию None).
    - size: размер точек на диаграмме рассеивания (по умолчанию 10).
    - grid: использовать сетку на диаграмме рассеивания (по умолчанию True).
    - figsize: размер диаграммы рассеивания (по умолчанию (15, 5)).
    - alpha: прозрачность точек на диаграмме рассеивания (по умолчанию 0.07).
    - ticklabel_format: формат отображения значений на оси (по умолчанию None).
    - trendline: отображение линии тренда (по умолчанию False).
    """
    
    if color == None:
        color = create_random_color()
    
    fig, ax = plt.subplots(figsize=figsize)
    ax.scatter(data[column_name_x], data[column_name_y], 
                color=color, s=size, alpha=alpha)
    
    plt.xlabel(xlabel if xlabel else column_name_x)
    plt.ylabel(ylabel if ylabel else column_name_y)
    plt.grid(grid)
    plt.title(title if title else 'диаграмма рассеивания')
    if ticklabel_format != None:
        try:
            plt.ticklabel_format(style=ticklabel_format)
        except:
            print("Try another format!")
            
    if trendline:
        slope, intercept = np.polyfit(data[column_name_x], data[column_name_y], 1)
        trendline = slope * np.array(data[column_name_x]) + intercept
        plt.plot(data[column_name_x], trendline, color=create_random_color())
```

##### create_pie_chart

```python
def create_pie_chart(data, column_name, title='', figsize=(15, 10), shadow=True):
    """
    Создает круговую диаграмму на основе указанного столбца из DataFrame.

    Параметры:
    - data: исходные данные.
    - column_name: наименование столбца данных.
    - title: заголовок диаграммы (по умолчанию пустая строка).
    - figsize: размер диаграммы рассеивания (по умолчанию (15, 10)).
    - shadow: наличие тени (по умолчанию True).
    """
    
    # создаем список цветов для будущей диаграммы
    # количество цветов = количеству уникальных значений в столбце
    colors = []
    for _ in range(data[column_name].nunique()):
        colors.append(create_random_color())
    # создаем рисунок
    fig, ax = plt.subplots(figsize=figsize)
    # настраиваем диаграмму
    # в качестве данных используем метод value_counts + сортировка sort_values
    # а в качестве лэйблов используем value_counts, sort_values и достаем индексы
    ax.pie(data[column_name].value_counts().sort_values(), autopct='%1.1f%%', 
           labels=data[column_name].value_counts().sort_values().index, 
           shadow=shadow, colors=colors, textprops={'fontsize': 14})
    
    ax.set_title(title if title else 'круговая диаграмма', fontsize=20)
```

#### Шаг 5. Подсчёт выручки

##### Создайте датафрейм с агрегированными данными о поездках на основе датафрейма с объединёнными данными из шага 4:
найдите суммарное расстояние, количество поездок и суммарное время для каждого пользователя за каждый месяц.

Чтобы выполнить эту задачу необходимо: 
- создать дополнительный столбик со значением месяца в df
- сгруппировать данные по  месяцу и id и добавить аггрегацию необходимым столбцам (aggregated_df)
- сбросим индексы и добавим необходимые столбцы из основной таблицы (df) для выполнения последующих задач (а именно подсчета выручки)

```python
df['month'] = df.date.dt.month
```

```python
aggregated_df = df.groupby(['user_id', 'month']).agg({
    'duration' : 'sum',
    'distance' : 'sum',
    'month' : 'count',
})
aggregated_df.rename(columns={'month' : 'count'}, inplace=True)
```

```python
aggregated_df.reset_index(inplace=True, level=['month', 'user_id'])
aggregated_df.head(20)
```

```python
aggregated_df = (aggregated_df.merge(df[['user_id','minute_price', 'start_ride_price', 
                                         'subscription_fee', 'subscription_type']], 
                                    on='user_id'))
```

```python
aggregated_df.head(20)
```

```python
aggregated_df.duplicated().sum()
```

```python
# после доавления среза столбцов в агригированную таблицу, появились дубликаты
# удаляем их 
# сбрасываем индексы
aggregated_df.drop_duplicates(inplace=True)
aggregated_df.reset_index(drop=True, inplace=True)
```

##### В этот же датафрейм добавьте столбец с помесячной выручкой, которую принёс каждый пользователь. 
Для этого обратитесь к информации об условиях оплаты для подписчиков и тех, у кого нет подписки. Продолжительность каждой поездки в каждой строке исходного датафрейма для подсчёта стоимости округляется до следующего целого числа: например, значения 
25.3, 25.5 и 26.0 должны быть преобразованы к 26.

Помесячная выручка считается по формуле: (стоимость старта поездки × количество поездок) + стоимость одной минуты поездки × общая продолжительность всех поездок в минутах) + стоимость подписки.

```python
aggregated_df['revenue'] = (((aggregated_df['start_ride_price'] * aggregated_df['count'])
                         + aggregated_df['minute_price'] * np.ceil(aggregated_df['duration']))
                            + aggregated_df['subscription_fee']) 
```

```python
aggregated_df.head(10)
```

##### removed_df подсчет выручки

```python
removed_df = pd.merge(pd.merge(removed_df, users_df, on='user_id'), 
                      subs_df, on='subscription_type')

removed_df['month'] = removed_df.date.dt.month
```

```python
agg_removed_df = removed_df.groupby(['month', 'user_id']).agg({'duration' : 'sum', 'distance' : 'sum','month' : 'count'})
agg_removed_df.rename(columns={'month' : 'count'}, inplace=True)
agg_removed_df.reset_index(inplace=True, level=['month', 'user_id'])

agg_removed_df = (agg_removed_df.merge(removed_df[['user_id','minute_price', 'start_ride_price', 
                                         'subscription_fee', 'subscription_type']], 
                                    on='user_id'))

agg_removed_df.drop_duplicates(inplace=True)
agg_removed_df.reset_index(drop=True, inplace=True)
```

```python
agg_removed_df.reset_index(inplace=True)
```

```python
# подсчет выручки
agg_removed_df['revenue'] = (((agg_removed_df['start_ride_price'] * agg_removed_df['count'])
                         + agg_removed_df['minute_price'] * np.ceil(agg_removed_df['duration']))
                            + agg_removed_df['subscription_fee']) 
```

##### Важно понять, тратят ли пользователи с подпиской больше времени на поездки? 
Если да, то пользователи с подпиской могут быть «выгоднее» для компании. Проверьте гипотезу. Используйте исходные данные о продолжительности каждой сессии — отдельно для подписчиков и тех, у кого нет подписки.

1. Сформулируем гипотезу:
    - Нулевая гипотеза (H0): Среднее время поездок пользователей с подпиской равно среднему времени поездок пользователей без подписки.
    - Альтернативная гипотеза (H1): Среднее время поездок пользователей с подпиской больше среднего времени поездок пользователей без подписки.

2. Выполнить тест:
    - Выполняем t-тест, используя функцию ttest_ind(). Установливаем параметр equal_var в False, чтобы указать, что мы считаем дисперсии выборок разными.
    - Получить p-value с помощью метода .pvalue.
    
3. Проанализировать результаты:
    - Если p-value меньше выбранного уровня значимости (alpha = 0.05), то отвергаем нулевую гипотезу и делаем вывод, что среднее время поездок пользователей с подпиской отличается от среднего времени поездок пользователей без подписки.
    - Если p-value больше выбранного уровня значимости, то нулевая гипотеза не может быть отвергнута и нельзя сделать вывод об отличии среднего времени поездок пользователей с подпиской от среднего времени поездок пользователей без подписки.


```python
# уровень значимости 5%
alpha = .05
# запишем результат метода ttest_ind в переменную results_duration,
# в качестве первого параментра укажем выборку пользователей с подпиской, а второго - выборка без подписки
# укажем параметр equal_var - False, 
# так как выборки у нас разного размера и скорее всего дисперсии у выборок отличаются
# укажем параметр alternative - greater для правосторонней гипотезы
results_duration = st.ttest_ind(ultra_df['duration'], free_df['duration'], 
                                equal_var=False, alternative='greater')
# ознакомимся с pvalue
print('p-значение:', results_duration.pvalue)
# сравниваем полученный pvalue с уровнем значимости (alpha)
# если pvalue меньше alpha, то нулевая гипотеза отклоняется
# иначе нулевая гипотеза принимается
if results_duration.pvalue < alpha:
    print('Разница во времени поездок пользователей с подпиской и без подписки статистически значима')
else:
    print('Нет статистически значимой разницы во времени')
```

На основании полученных результатов, можно сделать вывод о статистической значимости разницы времени поездок на самокатах пользователей с подпиской и без подписки. В результате применения метода ttest_ind из библиотеки scipy к двум выборкам, значения p_value получились меньше выбранного уровня значимости (0.05). Таким образом, на основании этих данных можем сделать вывод, что пользователи с подпиской тратят больше времени на поездки на самокатах.

##### Расстояние одной поездки в 3130 метров — оптимальное с точки зрения износа самоката. 
Можно ли сказать, что расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает 3130 метров? Проверьте гипотезу и сделайте выводы.

Для проверки данной гипотезы, используем одновыборочный T-тест (ttest_1samp).
1. Сформулируем гипотезу:
    - Пусть нулевая гипотеза (H0): - расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает 3130 метров.
    - Альтернативная гипотеза (H1): - расстояние, которое проезжают пользователи с подпиской за одну поездку, превышает 3130 метров.
2. Выполнить тест:
    - Определим наше оптимальное значение. 
    - Выполняем t-тест, используя функцию ttest_1samp().
    - Получить p-value с помощью метода .pvalue.
    
3. Проанализировать результаты:
    - Если p-value меньше выбранного уровня значимости (alpha = 0.05), то отвергаем нулевую гипотезу и делаем вывод, что среднее расстояние, которое преодолевает пользователь с подпиской, превышает 3130 метров.
    - Если p-value больше выбранного уровня значимости, то нулевая гипотеза не может быть отвергнута и нельзя сделать вывод, что среднее расстояние, которое преодолевает пользователь с подпиской, не превышает 3130 метров.

```python
# уровень значимости 5% и он определен в переменной alpha
# сохраним в переменную interested_value наше оптимальное значение
interested_value = 3130
# запишем результат метода ttest_1samp в переменную results_distance,
# в качестве параментра укажем выборку пользователей с подпиской
results_distance = st.ttest_1samp(ultra_df['distance'], interested_value, alternative='greater')
# ознакомимся с pvalue
print('p-значение:', results_distance.pvalue)
# сравниваем полученный pvalue с уровнем значимости (alpha)
# если pvalue меньше alpha, то нулевая гипотеза отклоняется
# иначе нулевая гипотеза принимается
if results_distance.pvalue < alpha:
    print('Отклоняем нулевую гипотезу: Среднее расстояние превышает 3130 метров')
else:
    print('Не отклоняем нулевую гипотезу: Среднее расстояние не превышает 3130 метров')
```

Итак, полученное p-value равно 0.0317, что меньше уровня значимости. Следовательно, у нас есть основания отвергать нулевую гипотезу. Мы можем сказать, что расстояние, которое проезжают пользователи с подпиской за одну поездку, превышает 3130 метров.

##### Проверьте гипотезу о том, будет ли помесячная выручка от пользователей с подпиской по месяцам выше, чем выручка от пользователей без подписки. 

Для проверки данной гипотезы, используем для независимых выборок (ttest_ind).
1. Сформулируем гипотезу:
    - Нулевая гипотеза (H0): Выручки от пользователей с подпиской и без неё равнозначны.
    - Альтернативная гипотеза (H1): Выручка от пользователей с подпиской по месяцам выше, чем выручка от пользователей без подписки.

2. Выполнить тест:
    - Выполняем t-тест, используя функцию ttest_ind(). Установливаем параметр equal_var в False, чтобы указать, что мы считаем дисперсии выборок разными.
    - Получить p-value с помощью метода .pvalue.
    
3. Проанализировать результаты:
    - Если p-value меньше выбранного уровня значимости (alpha = 0.05), то отвергаем нулевую гипотезу и делаем вывод, что выручка от пользователей с подпиской в среднем выше.
    - Если p-value больше выбранного уровня значимости, то нулевая гипотеза не может быть отвергнута и нельзя сделать вывод об отличии выручки от пользователей с подпиской и без неё.

```python
# уровень значимости 5% и он определен в переменной alpha

# запишем в переменную ultra_month_revenue выручку от пользователей с подпиской
ultra_month_revenue = aggregated_df[aggregated_df['subscription_type'] == 'ultra']['revenue']

# запишем в переменную free_month_revenue выручку от пользователей без подписки
free_month_revenue = aggregated_df[aggregated_df['subscription_type'] == 'free']['revenue']

# запишем результат метода ttest_ind в переменную results_distance,
# в качестве параментров укажем переменные ultra_month_revenue и free_month_revenue
# укажем параметр equal_var - False, 
# так как выборки у нас разного размера и скорее всего дисперсии у выборок отличаются
# укажем параметр alternative - greater для правосторонней гипотезы
results_revenue = st.ttest_ind(ultra_month_revenue, free_month_revenue, 
                               equal_var=False, alternative='greater')

# ознакомимся с pvalue
print('p-значение:', results_revenue.pvalue)
# сравниваем полученный pvalue с уровнем значимости (alpha)
# если pvalue меньше alpha, то нулевая гипотеза отклоняется
# иначе нулевая гипотеза принимается
if results_revenue.pvalue < alpha:
    print('Отклоняем нулевую гипотезу: есть статистически значимые различия в выручке между двумя группами пользователей по месяцам')
else:
    print('Не отклоняем нулевую гипотезу: у нас нет достаточных доказательств для отвержения нулевой гипотезы о равенстве выручки от пользователей с подпиской и без подписки.')
```

Результаты теста показали, что существуют статистически значимые различия в выручке между двумя группами пользователей по месяцам. Это означает, что на основе имеющихся данных, можно сделать вывод о том, что выручка от пользователей с подпиской в среднем выше, чем выручка от пользователей без подписки.

##### Представьте такую ситуацию: техническая команда сервиса обновила сервера, с которыми взаимодействует мобильное приложение. 
Она надеется, что из-за этого количество обращений в техподдержку значимо снизилось. Некоторый файл содержит для каждого пользователя данные о количестве обращений до обновления и после него. Какой тест вам понадобился бы для проверки этой гипотезы?

Ответ:
- Чтобы проверить гипотезу о значимом снижении количества обращений в техподдержку после обновления серверов, следует использовать <b>ttest_rel</b> для связанных выборок. В данном случае у нас есть данные о количестве обращений для каждого пользователя до обновления серверов и после него. Поскольку для одного и того же пользователя имеются связанные значения (до и после обновления), мы можем использовать ttest_rel для связанных выборок для оценки различий между двумя условиями.
