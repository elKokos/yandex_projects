##### describe_data

```python
def describe_data(data):
    """
     Выводит основные характеристики DataFrame, включая:
    - 10 случайных строк из таблицы
    - размерность DataFrame
    - информацию по числовым столбцам
    - информацию по строковым столбцам
    - количество пропусков в таблице
    - количество уникальных значений в каждом столбце
    - количество дублирующихся строк
    - количество уникальных значений в каждом столбце
    - гистограммы по всем столбцам
    - тепловую карту корреляции
    """
    # вывод 10 случайных строк
    print('\033[92mвывод случайных строк из таблицы:\033[90m\n')
    # если строк в таблице меньше 10, то выскочит ошибка
    # отлавливаем ее и выводим все имеющиеся строки в таблице
    try:
        display(data.sample(10))
    except:
        display(data.sample(data.shape[0]))
    
    # вывод типов данных в столбцах
    print('Типы данных:')
    print(data.dtypes)
    ('-----------------------------------------------------------------------')
    
    # вывод количества строк
    print('\033[92mКоличество строк:\033[90m', f'\033[91m{data.shape[0]}\033[90m')
    print('-----------------------------------------------------------------------')
    
    # вывод количества столбцов 
    print('\033[92mКоличество столбцов:\033[90m', f'\033[91m{data.shape[1]}\033[90m')
    print('-----------------------------------------------------------------------')
    
    # вывод информации о числовых столбцах
    print('\033[92mИнформация по числовым столбцам:\033[90m\n')
    for column in data.select_dtypes(include='number'):
        print(data[column].describe(), '\n')
        print('-----------------------------------------------------------------------')
    
    # вывод информации об оъектных данных
    print('\033[92mИнформация по строковым столбцам:\033[90m\n')
    for column in data.select_dtypes(include='object'):
        print(data[column].describe(), '\n')
        print('-----------------------------------------------------------------------')
    
    # вывод столбцов, содержащих пропуски, и их количество
    print('\033[92mКоличество пропусков в таблице:\033[90m\n')
    # эта переменная нужно для подсчета столбиков с пропусками
    count_na = 0
    for column in data.columns:
        null_in_column = data[column].isna().sum()
        
        if null_in_column:
            count_na += 1
            print(f'\033[91m{column}: {null_in_column}\033[90m')
            print('-----------------------------------------------------------------------')
    if count_na == 0:
        print('В таблице нет пропусков 🥂')
    
    # вывод информации об уникальных значениях каждого столбца
    print('\033[92mКоличество уникальных значений в каждом столбце:\033[90m\n')
    for column in data.columns:
        unique_values = data[column].nunique()
        print(f'{column} : {unique_values}')
        print('-----------------------------------------------------------------------')
    
    # вывод информации о дублирующих строках
    print('\n')
    duplicated_value = data.duplicated().sum()
    if duplicated_value:
        print(f'Количество дублирующихся строк: \033[91m{duplicated_value}\033[90m')
    else:
        print('\033[94mДублирующихся строк не обнаружено\033[90m')
    print('\n')
    
    # построение гистограммы по всем столбцам (для ознакомления)
    print('\n')
    print('\033[92mГистограммы по всем столбцам и тепловая карта корреляции:\033[90m\n')
    try:
        data.hist(figsize=(10, 10))
        sns.heatmap(data.corr(), annot=True)
    except:
        print('не получилось построить гистограммы')
    
```

##### create_missing_df

```python
def create_missing_df(data):
    """
    функция, которая сформирует нам таблицу с количеством пропусков и вернет ее
    на вход принимает только один параметр - DataFrame
    """
    # формируем дф с количеством пропусков в столбцах
    missing_df = data.isna().sum(axis=0).reset_index()
    missing_df.columns = ['column_name', 'missing_count']
    missing_df = missing_df[missing_df['missing_count'] > 0]
    missing_df = missing_df.sort_values('missing_count', ascending=True)
    # считаем % пропусков от общего
    missing_df['%'] = np.round((missing_df['missing_count'] / data.shape[0])*100, 2)
    return missing_df
```

##### remove_outliners_IQR

```python
def remove_outliners_IQR(data, column_name):
    """
    Функция использует межквартильный диапазон для удаления выбросов из DataFrame,
    создает графики до и после удаления выбросов,
    возвращает обновленный DataFrame и DataFrame с аномалиями.
    выводит на экран количество удаленных строк
    Параметры:
    - data: исходные данные.
    - column_name: наименование столбца данных.
    """
    # создаем копию таблицы
    data_after_remove = data.copy()
    # вычисляем статистические характеристики столбца (квартили)
    Q1 = data_after_remove[column_name].quantile(0.25)
    Q3 = data_after_remove[column_name].quantile(0.75)
    IQR = Q3 - Q1
    # вычисляем границы межквартильного диапазона
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # удаляем выбросы из столбца
    # запишем удаленные строчки в новую таблицу
    removed_df = data_after_remove[(data_after_remove[column_name] < lower_bound) | 
                                   (data_after_remove[column_name] > upper_bound)].copy()
    
    # перезапишем основную таблицу
    data_after_remove = data_after_remove[(data_after_remove[column_name] >= lower_bound) & 
                                          (data_after_remove[column_name] <= upper_bound)]
    
    # визуализируем графики до и после
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.boxplot(data[column_name])
    plt.title('До удаления выбросов')
    plt.ticklabel_format(style='plain', axis='y')
    
    plt.subplot(1, 2, 2)
    plt.boxplot(data_after_remove[column_name].dropna())
    plt.title('После удаления выбросов')
    plt.ticklabel_format(style='plain', axis='y')
    
    print(f'удалено {data.shape[0] - data_after_remove.shape[0]} строк')
    
    return data_after_remove, removed_df
```

##### create_random_color

```python
def create_random_color():
    """
    Функция генерирует случайный цвет 
    """
    r = random.random()
    g = random.random()
    b = random.random()
    return (r, g, b)
```

##### create_hist

```python
def create_hist(column_name, data, bins=10, 
                xlabel='', ylabel='', title='', 
                color=None, alpha=0.7, grid=True, 
                ticklabel_format=None, figsize=(10, 5)):
    """
    Функция, создает гистрограмму на основе указанного столбца из DataFrame.
    
    Параметры:
    - column_name: наименование столбца данных.
    - data: исходные данные.
    - bins: количество корзин (по умолчанию 10).
    - xlabel: подпись оси X (по умолчанию пустая строка).
    - ylabel: подпись оси Y (по умолчанию пустая строка).
    - title: заголовок гистограммы (по умолчанию пустая строка).
    - color: цвет столбцов гистограммы (по умолчанию None).
    - alpha: прозрачность столбцов гистограммы (по умолчанию 0.7).
    - grid: использовать сетку в гистограмме (по умолчанию True).
    - ticklabel_format: формат отображения значений на оси (по умолчанию None).
    - figsize: размер гистограммы (по умолчанию (10, 5)).
    """
    
    # если цвет не был определен при вызове функции, то создаем случайный цвет
    # иначе присваиваем цвет при вызове
    if color == None:
        color = create_random_color()
    
    # настраиваем гистограмму
    fig, ax = plt.subplots(figsize=figsize)
    ax.hist(data[column_name],
            bins=bins, 
            color=color, 
            alpha=alpha)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(grid)
    if ticklabel_format != None:
        try:
            plt.ticklabel_format(style=ticklabel_format)
        except:
            print("Try another format!")
```

##### create_scatter_plot

```python
def create_scatter_plot(column_name_x, column_name_y,
                        data, xlabel='', ylabel='', 
                        title='', color=None, size=10, grid=True, 
                        figsize=(15, 5), alpha=0.07, ticklabel_format=None, trendline=False):
    """
    Создает диаграмму рассеивания на основе указанных столбцов из DataFrame.

    Параметры:
    - column_name_x: имя столбца, данные которого будут отображены на оси x.
    - column_name_y: имя столбца, данные которого будут отображены на оси y.
    - data: исходные данные.
    - xlabel: подпись оси X (по умолчанию пустая строка).
    - ylabel: подпись оси Y (по умолчанию пустая строка).
    - title: заголовок диаграммы (по умолчанию пустая строка).
    - color: цвет точек на диаграмме рассеивания (по умолчанию None).
    - size: размер точек на диаграмме рассеивания (по умолчанию 10).
    - grid: использовать сетку на диаграмме рассеивания (по умолчанию True).
    - figsize: размер диаграммы рассеивания (по умолчанию (15, 5)).
    - alpha: прозрачность точек на диаграмме рассеивания (по умолчанию 0.07).
    - ticklabel_format: формат отображения значений на оси (по умолчанию None).
    - trendline: отображение линии тренда (по умолчанию False).
    """
    
    if color == None:
        color = create_random_color()
    
    fig, ax = plt.subplots(figsize=figsize)
    ax.scatter(data[column_name_x], data[column_name_y], 
                color=color, s=size, alpha=alpha)
    
    plt.xlabel(xlabel if xlabel else column_name_x)
    plt.ylabel(ylabel if ylabel else column_name_y)
    plt.grid(grid)
    plt.title(title if title else 'диаграмма рассеивания')
    if ticklabel_format != None:
        try:
            plt.ticklabel_format(style=ticklabel_format)
        except:
            print("Try another format!")
            
    if trendline:
        slope, intercept = np.polyfit(data[column_name_x], data[column_name_y], 1)
        trendline = slope * np.array(data[column_name_x]) + intercept
        plt.plot(data[column_name_x], trendline, color=create_random_color())
```

##### create_pie_chart

```python
def create_pie_chart(data, column_name, title='', figsize=(15, 10), shadow=True):
    """
    Создает круговую диаграмму на основе указанного столбца из DataFrame.

    Параметры:
    - data: исходные данные.
    - column_name: наименование столбца данных.
    - title: заголовок диаграммы (по умолчанию пустая строка).
    - figsize: размер диаграммы рассеивания (по умолчанию (15, 10)).
    - shadow: наличие тени (по умолчанию True).
    """
    
    # создаем список цветов для будущей диаграммы
    # количество цветов = количеству уникальных значений в столбце
    colors = []
    for _ in range(data[column_name].nunique()):
        colors.append(create_random_color())
    # создаем рисунок
    fig, ax = plt.subplots(figsize=figsize)
    # настраиваем диаграмму
    # в качестве данных используем метод value_counts + сортировка sort_values
    # а в качестве лэйблов используем value_counts, sort_values и достаем индексы
    ax.pie(data[column_name].value_counts().sort_values(), autopct='%1.1f%%', 
           labels=data[column_name].value_counts().sort_values().index, 
           shadow=shadow, colors=colors, textprops={'fontsize': 14})
    
    ax.set_title(title if title else 'круговая диаграмма', fontsize=20)
```

##### create_unique_barplot

```python
def create_unique_barplot(data):
    """
    Создает столбчатый график с количеством уникальных значений в каждом столбце на основе полученного DataFrame.

    Параметр:
    - data: исходные данные.
    """
    # создаем рисунок
    plt.figure(figsize=(16, 5))
    # список наименований стобцов
    cols = data.columns
    # список количества уникальных значений, списковое включение (List comprehension)
    uniques = [len(data[col].unique()) for col in cols]
    # создание столбчатого графика
    ax = sns.barplot(x=cols, y=uniques, palette='hls', log=True)
    # установка наименований для осей графика и наименование графика
    ax.set(xlabel='Наименования столбцов', ylabel='Количество', 
           title='Количество уникальных значений в каждом столбце')
    # количество значений для каждого столбца
    for p, uniq in zip(ax.patches, uniques):
        ax.text(p.get_x() + p.get_width()/2.,
                uniq + 10,
                uniq,
                ha="center") 

    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

##### create_top_10_unique

```python
def create_top_10_unique(data, column_name, xlabel='', ylabel='', title=''):
    """
    Функция создает топ 10 часто встречающихся уникальных значений
    Параметры:
    - data: исходные данные.
    - column_name: наименование столбца данных.
    - xlabel: подпись оси X (по умолчанию пустая строка).
    - ylabel: подпись оси Y (по умолчанию пустая строка).
    - title: заголовок диаграммы (по умолчанию пустая строка).
    """
    x=data[column_name].value_counts().iloc[:10].index
    y=data[column_name].value_counts().iloc[:10].values
    
    plt.figure(figsize=(16,5))

    ax = sns.barplot(x=x, y=y)
    ax.set_xlabel(xlabel if xlabel else column_name, fontsize=16)
    ax.set_ylabel(ylabel if ylabel else 'Количество', fontsize=16)
    ax.set_title(title if title else 'Топ 10 уникальных значений', fontsize=20)
    plt.xticks(rotation=40)
    
    # количество значений для каждого столбца
    for i in range(len(x)):
        plt.text(i, y[i] + 1, str(y[i]), ha='center')

    
```

##### Обзор на то, как менялись продажи в зависимости от платформы
Выберите платформы с наибольшими суммарными продажами и постройте распределение по годам. За какой характерный срок появляются новые и исчезают старые платформы?

```python
# создаем рисунок размером 20х10
fig, ax = plt.subplots(figsize=(20, 10))
# создание сводной таблицы, в которой индексы будут вид платформы
# колонки - года
# значения - суммарные продажи
# при создании графика добавим аргумент stacked=True для создания визуального накопительного эффекта
df.pivot_table(index='year_of_release',
              columns='platform',
              values='total_sales',
              aggfunc='count').plot(kind='bar',stacked=True, ax=ax, figsize=(16,9), 
                                    color=[create_random_color() for _ in range(df.platform.nunique())])

# стилистические настройки графика

# размещаем легенду за пределами графика
ax.legend(bbox_to_anchor=(1.0, 1.0), title='platform')
ax.set_xlabel('Годы', fontsize=20)
ax.set_ylabel('Количество проданных игр', fontsize=20)
ax.set_title('Распределение количества продаж компьютерных игр в разрезе платформ', fontsize=20)
ax.grid()
None
```

```python
# прежде чем вывести график по продажам по топ 10 платформам, сначала получим список 10 наиболее популярных платформ
top_10_platforms = (df[df['year_of_release'].between(2012, 2016)]
                    .groupby(['platform'])['total_sales'].sum()
                    .sort_values(ascending=False)[:10])

# создадим сводную таблицу по платформам и году, в качестве значений будут сумма проданных копий
platforms = (df.groupby(['platform', 'year_of_release'])['total_sales']
                    .sum()
                    .reset_index()
                    .sort_values(['platform', 'total_sales'], ascending=False))

# выделим из сводной таблицы только те строки, которые включают в себя платформу из топ 10
platforms = platforms[platforms['platform'].isin(top_10_platforms.index)]
# создаем график
plt.figure(figsize = (20, 8))
sns.lineplot(data = platforms, x='year_of_release', 
             y='total_sales', hue='platform', palette = "tab10", 
             linewidth = 3)

# стилизация графика

plt.title('Топ 10 продаж по платформам', fontsize=20)
plt.xlabel('Годы', fontsize=20)
plt.ylabel('Количество проданных игр', fontsize=20)
plt.xticks(platforms['year_of_release'][::1], rotation = 45)
# указываем период начиная с 1994
plt.xlim(1994, None)
plt.grid()
```

#### Проверка гипотез

- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Задайте самостоятельно пороговое значение alpha.

<a id='3_5_1'></a>
##### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

1. Сформулируем гипотезу:
    - Нулевая гипотеза (H0): Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
    - Альтернативная гипотеза (H1): Средние пользовательские рейтинги платформ Xbox One и PC различаются.

2. Выполнить тест:
    - Выполняем t-тест, используя функцию ttest_ind(). Установливаем параметр equal_var в False, чтобы указать, что мы считаем дисперсии выборок разными.
    - Получить p-value с помощью метода .pvalue.
    
3. Проанализировать результаты:
    - Если p-value меньше выбранного уровня значимости (alpha = 0.05), то отвергаем нулевую гипотезу и делаем вывод, что пользовательские рейтинги платформ Xbox One и PC различаются.
    - Если p-value больше выбранного уровня значимости, то нулевая гипотеза не может быть отвергнута и нельзя сделать вывод об отличии средних пользовательских рейтингов платформ Xbox One и PC.

Метод ttest_ind из библиотеки scipy не игнорирует пропуски при вычислении t-статистики и p-значения. Это означает, что при наличии пропусков, метод будет работать со значениями, которые также являются пропущенными. Поэтому удалим пропуски в данных.

```python
# уровень значимости 5%
alpha = .05
# запишем результат метода ttest_ind в переменную results_rating_xone_pc,
# в качестве первого параментра укажем выборку пользовательких рейтингов xone, 
# а второго - выборка пользовательких рейтингов pc
# удаляем у каждой выборки Nan
# укажем параметр equal_var - False, 
# так как выборки у нас разного размера и скорее всего дисперсии у выборок отличаются
results_rating_xone_pc = st.ttest_ind(xone_df.user_score.dropna().reset_index(drop=True), 
                                      pc_df.user_score.dropna().reset_index(drop=True), 
                                      equal_var=False)

# ознакомимся с pvalue
print('p-значение:', results_rating_xone_pc.pvalue)
# сравниваем полученный pvalue с уровнем значимости (alpha)
# если pvalue меньше alpha, то нулевая гипотеза отклоняется
# иначе нулевая гипотеза принимается
if results_rating_xone_pc.pvalue < alpha:
    print('Разница в средних пользовательских рейтингах платформ Xbox One и PC статистически значима')
else:
    print('Нет статистически значимой разницы в средних пользовательских рейтингах платформ Xbox One и PC')
```

Для проверки гипотезы о равенстве средних мы использовали модифицированный t-критерий Уэлча (Welch's t-test). Этот критерий позволяет учитывать различия в дисперсиях между выборками.

Вывод: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые и нулевая гипотеза не отклоняется.

##### Средние пользовательские рейтинги жанров Action и Sports разные

1. Сформулируем гипотезу:
    - Нулевая гипотеза (H0): Средние пользовательские рейтинги жанров Action и Sports одинаковые.
    - Альтернативная гипотеза (H1): Средние пользовательские рейтинги жанров Action и Sports разные.

2. Выполнить тест:
    - Выполняем t-тест, используя функцию ttest_ind(). Установливаем параметр equal_var в False, чтобы указать, что мы считаем дисперсии выборок разными.
    - Получить p-value с помощью метода .pvalue.
    
3. Проанализировать результаты:
    - Если p-value меньше выбранного уровня значимости (alpha = 0.05), то отвергаем нулевую гипотезу и делаем вывод, что пользовательские рейтинги жанров Action и Sports различаются.
    - Если p-value больше выбранного уровня значимости, то нулевая гипотеза не может быть отвергнута и нельзя сделать вывод об отличии средних пользовательских рейтингов жанров Action и Sports.

```python
# создадим две таблицы с двумя жанрами action и sports
action_df = (df[(df['year_of_release'] >= 2012) 
                          & (df['genre'].isin(['action']))])

sports_df = (df[(df['year_of_release'] >= 2012) 
                          & (df['genre'].isin(['sports']))])
```

```python
# уровень значимости 5% и он определен в переменной alpha

# запишем результат метода ttest_ind в переменную results_rating_action_sports,
# в качестве первого параментра укажем выборку пользовательких рейтингов жанра action, 
# а второго - выборка пользовательких рейтингов sports
# удаляем у каждой выборки Nan
# укажем параметр equal_var - False, 
# так как выборки у нас разного размера и скорее всего дисперсии у выборок отличаются
results_rating_action_sports = st.ttest_ind(action_df.user_score.dropna().reset_index(drop=True), 
                                      sports_df.user_score.dropna().reset_index(drop=True), 
                                      equal_var=False)

# ознакомимся с pvalue
print('p-значение:', results_rating_action_sports.pvalue)
# сравниваем полученный pvalue с уровнем значимости (alpha)
# если pvalue меньше alpha, то нулевая гипотеза отклоняется
# иначе нулевая гипотеза принимается
if results_rating_action_sports.pvalue < alpha:
    print('Разница в средних пользовательских рейтингах жанров Action и Sports статистически значима')
else:
    print('Нет статистически значимой разницы в средних пользовательских рейтингах жанров Action и Sports')
```

Вывод: Средние пользовательские рейтинги жанров Action и Sports разные
